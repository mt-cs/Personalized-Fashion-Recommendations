```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pandasql import sqldf

df_article = pd.read_csv('h-and-m-personalized-fashion-recommendations/articles.csv')
df_article
```

```python
df_article.columns
```

```python
df_art = df_article

df_art.drop(['product_code', 'prod_name', 'product_type_no',
       'product_type_name', 'graphical_appearance_no', 'colour_group_code', 'colour_group_name',
       'perceived_colour_value_id', 'perceived_colour_master_id', 'perceived_colour_master_name',
       'department_no', 'department_name', 'index_code', 'index_name',
       'index_group_no', 'section_no', 'garment_group_no', 'detail_desc'], inplace=True, axis=1)

df_art
```

```python
df_art['product_group_name'].value_counts()
```

```python
df_art['section_name'].value_counts()
```

```python
df_art['garment_group_name'].value_counts()
```

```python
temp = df_art['graphical_appearance_name'].value_counts() <= 2000
temp = temp.loc[temp]
mask = set(temp.index)
df_art["modified_graphical_appearance"] = df_art["graphical_appearance_name"].apply(lambda x : 'other' if x in mask else x)
```

```python
df_art['modified_graphical_appearance'].value_counts()
```

```python
temp = df_art['product_group_name'].value_counts() <= 2000
temp = temp.loc[temp]
mask = set(temp.index)
df_art["modified_product_group"] = df_art["product_group_name"].apply(lambda x : 'other' if x in mask else x)

df_art['modified_product_group'].value_counts()
```

```python
temp = df_art['section_name'].value_counts() <= 3328
temp = temp.loc[temp]
mask = set(temp.index)
df_art["modified_section"] = df_art["section_name"].apply(lambda x : 'other' if x in mask else x)

df_art['modified_section'].value_counts()
```

```python
temp = df_art['garment_group_name'].value_counts() <= 4874
temp = temp.loc[temp]
mask = set(temp.index)
df_art["modified_garment_group"] = df_art["garment_group_name"].apply(lambda x : 'other' if x in mask else x)

df_art['modified_garment_group'].value_counts()
```

```python
df_art.columns
```

```python
df_art.drop(['graphical_appearance_name', 'product_group_name', 'section_name', 'garment_group_name'], inplace=True, axis=1)

df_art
```

# CUSTOMER

```python
df_customer = pd.read_csv('h-and-m-personalized-fashion-recommendations/customers.csv')
df_customer
```

```python
df_customer.columns
```

```python
df_cust = df_customer

df_cust.drop(['FN', 'fashion_news_frequency', 'postal_code'], inplace=True, axis=1)

df_cust
```

```python
df_customer[['club_member_status']] = df_customer[['club_member_status']].fillna('N/A')
df_cust[['club_member_status']] = df_cust[['club_member_status']].fillna('N/A')
```

```python
df_customer[['Active']] = df_customer[['Active']].fillna(0)
df_cust[['Active']] = df_cust[['Active']].fillna(0)
```

# TRANSACTION

```python
df_transaction = pd.read_csv('h-and-m-personalized-fashion-recommendations/transactions_train.csv')
df_transaction
```

```python
df_transaction[["day", "month", "year"]] = df_transaction["t_dat"].str.split("-", expand = True)
df_transaction
```

```python
df_transaction.drop(['t_dat'], inplace=True, axis=1)
```

```python
df_cust_trans = df_transaction.merge(df_cust, on=["customer_id"], how="inner")
df_cust_trans
```
