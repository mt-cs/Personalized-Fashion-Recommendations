### Table Information Summary:

As it can be seen above, the table has 105542 rows (articles) and 25 columns (features) from which 11 have numeric data type (int64) and  14 string data type (object). [ROBERT KWIATKOWSKI](https://www.kaggle.com/datark1/detailed-eda-understanding-h-m-data) has a thorough explanation of the column that helped us understand them better. Here are a brief explanation of each column:

- **article_id (int64):** The unique ID for each article
- **product_code (int64):** A 6-digit code for each product
- **prod_name (object):** Product name
- **product_type_no (int64):** product type number
- **product_type_name (object):** Product type name
- **product_group_name (object):** Product group name
- **graphical_appearance_no (int64):** Product pattern code
- **graphical_appearance_name (object):** Product pattern name
- **colour_group_code (int64):** Code for the color group (50 major colour groups)
- **colour_group_name (object):** Name of the color group 
- **perceived_colour_value_id (int64):** perceived colour id (a broader grouping of the product colours with 8 categories)
- **perceived_colour_value_name (object):** Perceived colour name
- **perceived_colour_master_id (int64):** Colour grouping with 20 categories 
- **perceived_colour_master_name (object):** Perceived colour master name
- **department_no (int64):** Department number with 299 unique values
- **department_name (object):** Department name
- **index_code (int64):** A product category with 10 unique values
- **index_name (object):** Product index name 
- **index_group_no (int64):** A product category with 5 unique values
- **index_group_name (object):** Index group name
- **section_no (int64):** Product section number
- **section_name (object):** Product section name
- **garment_group_no (int64):** Garment group number
- **garment_group_name (object):** Garment group name
- **detail_desc (object):** Detailed description of the product

### Articles Types:
It would be informative to look at the distribution of index groups to see what categories the mojority of our articles fall into. 

```python
values = df['index_group_name'].value_counts().tolist()
labels = df['index_group_name'].unique().tolist()
plt.figure(figsize=(8, 6))
plt.title("Number of articles per index group")
prod_grp_name = pd.DataFrame({"Index Group Name":labels, "Number of Articles":values})
prod_grp_name = prod_grp_name.sort_values(["Number of Articles"], ascending=False)
s1 = sns.barplot(x = 'Index Group Name', y = 'Number of Articles', data = prod_grp_name, color = 'blue')
s1.set_xticklabels(s1.get_xticklabels(), rotation=90)
locs, labels = plt.xticks()

total = len(df)
for p in s1.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width() / 2 - 0.05
    y = p.get_y() + p.get_height()
    s1.annotate(percentage, (x, y), size = 12)
plt.show()
```


As it can be seen in the above chart, the main article categories are Ledieswear, Baby/Children, Menswear, sport, and divided. The majority of our articles fall into Ladieswear category. The second major category is Baby/children. This shows that there is more diversity among ladies and children articles compared to men's, sport's, and divided articles.

Now let us see what are the subcategories of the index group names:

***
To have a better understanding of the Distribution of major categories and their subcategories, we can show them on a barchart:

```python
gr_df = df.groupby(['index_group_name', 'index_name'])["article_id"].nunique().reset_index()

sns.set(style='white')

p = sns.catplot(x="index_group_name", y="article_id", hue="index_name", kind="bar", data=gr_df)
p.fig.set_size_inches(15, 8)
p.fig.subplots_adjust(top=0.81, right=0.86)

plt.title('Number of articles per index group', fontsize=16)
plt.xlabel('Index Group Name')
plt.ylabel('Number of Articles')

#rotate x-axis labels
plt.xticks(rotation=45)

ax = p.facet_axis(0, 0)

# iterate through the axes containers
for c in ax.containers:
    labels = [f'{(v.get_height() / 1000):.1f}K' for v in c]
    ax.bar_label(c, labels=labels, label_type='edge', fontsize = 7)

```

Notice that Baby/Children is devided into 4 different subcategories namely, Baby Sizes 50-98, Children Accessories and Swimwear, Children sizes 134-170, and Children Sizes 92-140. Furthermore, Ladieswear is subcategorized as Ladies Accessories, Ladieswear, and Lingeries/Tights.
